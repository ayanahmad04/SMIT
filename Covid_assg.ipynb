# 4. Create Function that take dataset columns and replace whitespace  with Underscore and update dataset automatically and the pass the dataset.

```python
def func(data):
    data.columns=data.columns.str.replace(' ','_').str.replace('/','_')
    for column in data.columns:
        if data[column].dtype == 'object':
            data[column] = data[column].str.replace(' ', '_')
    return data
d=func(data)
d.sample(6)
```
